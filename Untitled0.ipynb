```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
try:
    df = pd.read_csv('/content//content/Titanic-Dataset (3).csv')
except FileNotFoundError:
    print("Error: 'your_dataset.csv' not found. Please replace it with your data file.")
    data = {'col1': [1, 2, 3, np.nan, 5],
            'col2': ['A', 'B', 'C', 'D', 'E'],
            'col3': [10.1, 20.2, np.nan, 40.4, 50.5]}
    df = pd.DataFrame(data)
print("First 5 rows of the DataFrame:")
display(df.head())
print("\nDataFrame Info:")
display(df.info())
print("\nDescriptive Statistics:")
display(df.describe())
print("\nMissing values per column:")
display(df.isnull().sum())
for col in df.select_dtypes(include=np.number).columns:
    df[col].fillna(df[col].mean(), inplace=True)
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nDataFrame after handling missing values:")
display(df.isnull().sum())
print("\nNumber of duplicate rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

print("Number of duplicate rows after removal:", df.duplicated().sum())
for col in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Box plot of {col}')
    plt.show()
for col in df.select_dtypes(include=np.number).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print("\nDataFrame shape after potential outlier removal:", df.shape)
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = MinMaxScaler()
numerical_cols = df.select_dtypes(include=np.number).columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print("\nDataFrame after feature scaling:")
display(df.head())

df = pd.get_dummies(df, columns=df.select_dtypes(include='object').columns, drop_first=True)

print("\nDataFrame after one-hot encoding:")
display(df.head())

print("\nDataFrame after potential feature engineering:")
display(df.head())
```
